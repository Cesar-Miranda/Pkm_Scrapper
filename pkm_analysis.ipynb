```python
import pandas as pd

# importing csv's from a github repository 
types = pd.read_csv("https://raw.githubusercontent.com/veekun/pokedex/master/pokedex/data/csv/types.csv")
type_efficacy = pd.read_csv("https://raw.githubusercontent.com/veekun/pokedex/master/pokedex/data/csv/type_efficacy.csv")

# importing from scrapped data
pkmns_name_type = pd.read_csv("pkmns_name_type.csv")
pkmns_stat = pd.read_csv("pkmns_stat.csv")
```

```python
# lets take a quickly look on the data structure
display(types, type_efficacy, pkmns_name_type, pkmns_stat)
```

```python
# Since our intention is to make a type-stat analysis for competitive game, we will get rid off some 
# kind of pokemons (megas, primal etc..., that tends to be left behind generation after generation)
# Fortunelly, the method will do part of this job for us since the type column doesn't have those pkm
# variations

pkmns_type_name_stat = pd.merge(pkmns_name_type, pkmns_stat, on="name")
```

```python
pkmns_type_name_stat.head(30)
```

```python
# interesting to look
pkmns_type_name_stat.describe()
```

```python
pkmns_type_name_stat.value_counts()
```

```python
print(pkmns_type_name_stat.dtypes)
```

```python
pkmns_type_name_stat["total_stat"] = pkmns_stat.drop(['name', 'number'],axis=1).sum(axis=1)
```

```python
display(pkmns_type_name_stat)
```

```python
# 600+ will get out
pkmns_type_name_stat.drop(pkmns_type_name_stat[pkmns_type_name_stat['total_stat'] > 600].index, inplace=True)
```

```python
# Now, our max total_stat is 600
pkmns_type_name_stat.describe()

# another cool thing to start looking for is the mean status of pokemon. 68 for hp, 78 for atk, 73 for def, 71 for sp.atk,
# 70 for sp. def and 67 for speed
```

```python
# It's also necessary to get all 'mega' pokemons out, because them are usually desconsidered from game to game and don't have 
# enough solid presence to be relevant to our analysis. We will also discard partner variants

discard = ["Mega ", "Partner"]
pkmns_type_name_stat.drop(pkmns_type_name_stat[pkmns_type_name_stat.name.str.contains('|'.join(discard))].index, inplace=True)
```

```python
# now we only have relevant pokemons for our analysis
pkmns_type_name_stat.head(30)
```

```python
# now we can do specif analysis by type like this one:
print('[Electric Type Data]')
display(pkmns_type_name_stat[pkmns_type_name_stat['type1'] == 'Electric'].describe())
# we can already see that Pokemons that have Electric type as their type1, in mean, have higher sp.def than physical defense
# and more sp.atk than physical atk, their lower stat tends to be hp, so they seem to be fragile and have high speed stats
```

```python
# now, the direction our analysis will go through is: are those status good for an electric type? Let's take a look on two
# elemental types that electric type have weakness against groun and rock
print('[Rock Type Data]')
display(pkmns_type_name_stat[pkmns_type_name_stat['type1'] == 'Rock'].describe())

print('\n \n [Ground Type Data]')
display(pkmns_type_name_stat[pkmns_type_name_stat['type1'] == 'Ground'].describe())
```

```python
# Rock types have higher mean atk and lower sp.atk, so the higher stat in sp.def of electrics aren't that convenient
# But Rock have really low sp.def, what can give some advantage to electric types, their hp also is a bit under the total
# mean of all types, their speed are also low, what means to say that probably an electric type would have a first chance to atk
# but, taking in consideration the high atk of Rock and low def of electric, that one first atk would have a high cost

# Ground situation is pretty similar to Rock, but with even another advantage, taking the efficacy table we would know that ground
# is immune to electric type atks, so the electric pokemon would lost the possibilie to use its strongest type moves.

# Now lets take a look on two other types that are weak againts electric types to understand the overall electric situation
print('[Flying Type Data]')
display(pkmns_type_name_stat[pkmns_type_name_stat['type1'] == 'Flying'].describe())
print('\n \n [Water Type Data]')
display(pkmns_type_name_stat[pkmns_type_name_stat['type1'] == 'Water'].describe())
```

```python
# lets bring the mean values closer in the roll bar
pkmns_type_name_stat.mean()

# So, Flying types have, in mean, higher sp. def than physical def. This is bad news for electric types. For water the difference
# is little. electric types have a mean of 70 in sp.atk and Flying/Water have the same value. So Electric types are lucky for
# having the type advantage
# for atk and sp. atk, water and flying are also balanced, the physical atk is even higher than Electric's physicial defense
# what gives some vantage to them.
# but, are they able to move first?
# water speed mean is 65 points and flying 87, while electric have a 86 speed status, what means that they keep their advantage
# against water but, flying could do some turning tables over electric.
```

```python
# So, this what we gonna do, but in a scalable way. Case to case, considering not only type 1 but also type 2, searching for
# the natural best pokemons, considering their resistances, weakness, immunities and the relation with their status.
# but first, we need to put our other tables, types / type_efficacy,  together 
# looking on their fields we can see that we have the id_structure, so, they are combined by that column, we will merge them, but
# first we need to get rid off the columns we don't need
types.drop('generation_id', axis=1, inplace=True)
types.drop('damage_class_id', axis=1, inplace=True)
print('Types columns')
types.head()
```

```python
# let's bring type_efficacy structure closer
type_efficacy.head()
# we're able to see that damage_types are the ones attacking and target_type are the attacked. Damage factor will tell us about
# effectiveness
# Now we need to connect the identifier with the id values
```

```python
display(type_efficacy, types)
```

```python
# type_effiency = pd.merge(type_efficacy, types, on="id")
damage_type = types[:]
damage_type.rename(columns = {'id':'damage_type_id'}, inplace = True)
print('[Damage Type]')
display(damage_type)

target_type = types[:]
target_type.rename(columns = {'id': 'target_type_id'}, inplace = True)
print('\n \n [Target Type]')
display(target_type)
```

```python
damage_efficiency = pd.merge(damage_type, type_efficacy, on="damage_type_id")
damage_efficiency = pd.merge(target_type, damage_efficiency, on="target_type_id")
damage_efficiency.rename(columns = {'identifier_x': 'target_type'}, inplace = True)
damage_efficiency.rename(columns = {'identifier_y': 'damage_type'}, inplace = True)
damage_efficiency.drop('damage_type_id', axis=1, inplace=True)
damage_efficiency.drop('target_type_id', axis=1, inplace=True)
display(damage_efficiency)

# now we have a unified table that fits our purpose
```
